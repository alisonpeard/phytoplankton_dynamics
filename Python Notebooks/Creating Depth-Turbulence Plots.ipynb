```python
import numpy as np
import equations
from equations import get_stationary
from equations import get_time_evolution
from equations import get_R
import matplotlib.pyplot as plt

# inital parameters
zmax = 30
tmax = 100
z_steps = int(zmax*10) # dz = 0.1
column_depth = np.arange(1,zmax+1,1)

# Define results matrices
stationary_A = np.zeros((z_steps,zmax)) 
stationary_Rb = np.zeros((z_steps,zmax))
stationary_Rd = np.zeros((z_steps,zmax))
stationary_Rs = np.zeros(zmax)
stationary_Rb.shape
```

# Turbulence-Depth Plots
## d=1.0

```python
d = 1.0
I_0 = 300.0

i = 0
for depth in column_depth:
    depth_steps = depth*10 # because dz=0.1 there are 10 times more steps than the depth
    stationary_A[:depth_steps,i], stationary_Rb[:depth_steps,i], stationary_Rd[:depth_steps,i], stationary_Rs[i] = get_stationary(zmax=depth,tmax=tmax, d=d, I0=I_0)
    i += 1
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.DataFrame(stationary_A).head()
```

```python
# Heatmap
plt.imshow(stationary_A[:,:],aspect=1,origin="lower",extent=(0,zmax,0,zmax))
plt.xlabel('water column depth (m)')
plt.ylabel('water depth (m)')
plt.title('A concentration for d = {}'.format(d))
plt.gca().invert_yaxis()
plt.colorbar()
```

### Create Visuals for R

```python
R = get_R(Rd,Rb)
```

```python
# Heatmap
plt.imshow(R[:,:],aspect=1,origin="lower",extent=(0,zmax,0,zmax),vmax=50)
plt.xlabel('water column depth (m)')
plt.ylabel('water depth (m)')
plt.title('R concentration for d = {}'.format(d))
plt.gca().invert_yaxis()
plt.colorbar()
```

```python
# save everything to results dataframes to save repeating
pd.DataFrame(stationary_A).to_csv("A_d1.csv")
pd.DataFrame(stationary_Rb).to_csv("Rb_d1.csv")
pd.DataFrame(stationary_Rd).to_csv("Rd_d1.csv")
pd.DataFrame(stationary_Rs).to_csv("Rs_d1.csv")
pd.DataFrame(R).to_csv("R_d1.csv")
```

# Time-Evolution Plots

```python
from mpl_toolkits.mplot3d import axes3d
```

## d = 10

```python
d=10.0
A_d10, Rb_d10, Rd_d10, Rs_d10, z_grid, time_steps, Nz, Nt = get_time_evolution(zmax=zmax, tmax=tmax, d=d, I0=I_0)
```

```python
# Visualize Shift
plt.plot(z_grid,A_d10[:,-1])
plt.plot(z_grid,A_d10[:,0],color = 'r')
plt.title('A shift for d = {}'.format(d))
```

```python
# Heatmap
plt.imshow(A_d10[:,:],aspect=tmax/zmax,origin="lower",extent=(0,tmax,0,zmax))
plt.xlabel('days')
plt.ylabel('depth (meters)')
plt.title('A concentration for d = {}'.format(d))
plt.gca().invert_yaxis()
plt.colorbar()
```

```python
#%matplotlib notebook

days = np.linspace(0,tmax,Nt)
T,Z = np.meshgrid(days,z_grid)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T,Z, A_d10, cmap='cool')
ax.set_xlabel('time (days)')
ax.set_ylabel('z (meters)')
ax.set_zlabel('C')
ax.set_title('A - plankton for d = {}'.format(d))
```
