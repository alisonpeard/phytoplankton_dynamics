```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Plots for d=0.1

```python
A_01 = pd.read_csv("Results DataFrames/A_d01.csv",header=None).iloc[1:,1:]
Rb_01 = pd.read_csv("Results DataFrames/Rb_d01.csv",header=None).iloc[1:,1:]
Rd_01 = pd.read_csv("Results DataFrames/Rd_d01.csv",header=None).iloc[1:,1:]

A_01 = A_01.to_numpy()
Rb_01 = Rb_01.to_numpy()
Rd_01 = Rd_01.to_numpy()

tmax = 100
zmax = A_01.shape[1]

# iterate through rows
R_01 = np.zeros(Rd_01.shape)
for i in range(len(Rd_01)):
   # iterate through columns
   for j in range(len(Rd_01[0])):
        R_01[i,j] = Rd_01[i,j] + Rb_01[i,j]
```

```python
plt.imshow(A_01[:,:],aspect=tmax/zmax,origin="lower",extent=(0,tmax,0,zmax))
plt.title("A concentration for d=0.1")
plt.xlabel('days')
plt.ylabel('depth (meters)')
plt.gca().invert_yaxis()
plt.colorbar()
```

```python
plt.imshow(R_01[:,:],aspect=tmax/zmax,origin="lower",extent=(0,tmax,0,zmax))
plt.title("R concentration for d=0.1")
plt.xlabel('days')
plt.ylabel('depth (meters)')
plt.gca().invert_yaxis()
plt.colorbar()
```
